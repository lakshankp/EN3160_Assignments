### Question 1

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python

intensity_transform = np.concatenate((
    np.linspace(0, 50, 50),       
    np.linspace(100, 255, 100),    
    np.linspace(150, 255, 106)     
), axis=0).astype(np.uint8)

# Read the grayscale image
input_image = cv.imread('a1images/a1images/emma.jpg', cv.IMREAD_GRAYSCALE)
transformed_image = cv.LUT(input_image, intensity_transform)
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Display original image
axes[0].imshow(input_image, cmap="gray", vmin=0, vmax=255)
axes[0].get_xaxis().set_visible(False)
axes[0].get_yaxis().set_visible(False)
axes[0].set_title("Original Image")

# Display intensity transformation function
axes[1].plot(intensity_transform)
axes[1].set_title("Intensity Transformation Function")
axes[1].set_xlabel("Input Intensity")
axes[1].set_ylabel("Output Intensity")
axes[1].grid(True)

# Display transformed image
axes[2].imshow(transformed_image, cmap="gray", vmin=0, vmax=255)
axes[2].get_xaxis().set_visible(False)
axes[2].get_yaxis().set_visible(False)
axes[2].set_title("Transformed Image")

plt.show()
```

### Question 2

#### a)

```python
brain_image = cv.imread('a1images/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

white_matter_transform = np.concatenate((
    np.linspace(0, 0, 120),      
    np.linspace(190, 210, 60),   
    np.linspace(0, 0, 76)        
), axis=0).astype(np.uint8)


white_matter_enhanced = cv.LUT(brain_image, white_matter_transform)


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(brain_image, cmap="gray", vmin=0, vmax=255)
axes[0].set_title("Original Brain MRI")
axes[0].set_axis_off()

axes[1].plot(white_matter_transform, color='blue')
axes[1].set_title("White Matter Enhancement Function")
axes[1].set_xlabel("Input Intensity")
axes[1].set_ylabel("Output Intensity")
axes[1].grid(True)
axes[1].set_xlim([0, 255])
axes[1].set_ylim([0, 255])

axes[2].imshow(white_matter_enhanced, cmap="gray", vmin=0, vmax=255)
axes[2].set_title("White Matter Enhanced")
axes[2].set_axis_off()

plt.tight_layout()
plt.show()
```

#### b)

```python
# Create intensity transformation to highlight gray matter
# Gray matter typically appears in higher intensities (180-255) in MRI
gray_matter_transform = np.concatenate((
    np.linspace(0, 0, 180),      # Set intensities 0-179 to black
    np.linspace(175, 240, 76)    # Map intensities 180-255 to enhanced contrast
), axis=0).astype(np.uint8)

# Apply the transformation to highlight gray matter using the same brain image from part a
gray_matter_enhanced = cv.LUT(brain_image, gray_matter_transform)

# Create figure with three subplots matching the style from part a
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot 1: Original MRI image (reusing the same brain image)
axes[0].imshow(brain_image, cmap="gray", vmin=0, vmax=255)
axes[0].set_title("Original Brain MRI")
axes[0].set_axis_off()

# Plot 2: Intensity transformation function (using red to distinguish from white matter plot)
axes[1].plot(gray_matter_transform, color='red')
axes[1].set_title("Gray Matter Enhancement Function")
axes[1].set_xlabel("Input Intensity")
axes[1].set_ylabel("Output Intensity")
axes[1].grid(True)
axes[1].set_xlim([0, 255])
axes[1].set_ylim([0, 255])

# Plot 3: Enhanced gray matter image
axes[2].imshow(gray_matter_enhanced, cmap="gray", vmin=0, vmax=255)
axes[2].set_title("Gray Matter Enhanced")
axes[2].set_axis_off()

# Adjust layout and display
plt.tight_layout()
plt.show()
```
